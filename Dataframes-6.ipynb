```python
certified, denied = [], []
for i in continents:
    con1 = df['continent'] == i
    con2 = df['case_status'] == 'Certified'
    con3 = df['case_status'] == 'Denied'
    certi_con = con1 & con2
    deni_con = con1 & con3
    certified.append(len(df[certi_con]))
    denied.append(len(df[deni_con]))
```

```python
conti_labels = df['continent'].unique()
case_labels = sorted(df['case_status'].unique())
pd.DataFrame(zip(certified, denied), index = conti_labels, columns = case_labels)
```

### Chart purposes

- bar chart: frequency table

- pie chart: relative frequency table

- histogram: frequency distribution plot

- box plot: Outlier analysis

- scatterplot: Relation between two numerical columns

### scatter plot

```python
x = list(range(1,11))
y = list(range(1,11))
plt.scatter(x,y)
plt.plot(x,y)
```

```python
x = list(range(-4,5))
y = [i*i for i in x]
plt.scatter(x,y)
plt.plot(x,y)
```

```python
col1 = df['prevailing_wage']
col2 = df['no_of_employees']
col3 = df['yr_of_estab']
plt.figure(figsize = (20,20))
plt.subplot(3,1,1).scatter(col1,col2)
plt.title('prevailing wage vs number of employees')
plt.subplot(3,1,2).scatter(col1,col3)
plt.title('prevailing wage vs year of establishment')
plt.subplot(3,1,3).scatter(col2,col3)
plt.title('number of emplyees vs year of establishment')
```
