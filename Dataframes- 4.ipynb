```python
df = pd.read_csv("C:/Users/ameen/Downloads/Visadataset - Visadataset.csv")
```

```python
cat = df.select_dtypes(include = 'object').columns
num = df.select_dtypes(exclude = 'object').columns
```

```python
cat = list(cat)
```

```python
import os
folder_name = 'bar_charts'
try:
    os.mkdir(folder_name)
except Exception as e:
    print(e)
```

```python
import os
folder_name = 'bar_charts'
try:
    os.mkdir(folder_name)
except Exception as e:
    print(e)

cwd = os.getcwd()
new_dir = os.path.join(cwd,folder_name)

for i in cat[1:]:
    plt.figure(figsize = (15,6))
    keys = df[i].value_counts().keys()
    values = df[i].value_counts().values
    file_name = f'{i}_bar_.jpg'
    path = os.path.join(new_dir, file_name)
    plt.bar(keys, values)
    plt.savefig(path)
```
