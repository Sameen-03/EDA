- One hot encoder creates new columns that equal to number of unique labels

- One disadvantage is dimensions will be increased

- More dimensions lead to more processing time, leads to bigger infra memory

- This is known as curse of dimensionality

- Generally ML algorithms expect no corelation between input columns

- When one column value is available and at the same time another column value is not available, it means both are independent of each other

```python
cat = cat.columns
```

```python
pd.get_dummies(df,columns = cat[1:], dtype = int)
```

- If data is ordinal, then go with label encoder
  
- If data is nominal, then go with one hot encoder

- If labels are more, go with Label encoder

- IF labels are less, go with one hot encoder
